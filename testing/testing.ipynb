### Basic Function: Replace inf with NaN and Drop All NaN Rows

```python
def replace_inf_and_drop_na(df_testing):
    """
    Replace infinite values with NaN and drop all rows containing NaN values.

    Parameters:
    df (pd.DataFrame): The DataFrame to process.

    Returns:
    None: The function modifies the DataFrame in place.
    """
    df_testing.replace([np.inf, -np.inf], np.nan, inplace=True)
    df_testing.dropna(inplace=True)
```

### Scale as a function

```python
def scale_and_concat(df_features, numerical_columns, categorical_columns):
    """
    Scales numerical columns and concatenates them with categorical columns.

    Parameters:
    df_scaled (np.array or pd.DataFrame): Scaled numerical data.
    df_features (pd.DataFrame): Original DataFrame containing the categorical columns.
    numerical_columns (list): List of numerical column names.
    categorical_columns (list): List of categorical column names.

    Returns:
    pd.DataFrame: DataFrame with scaled numerical features and original categorical features.
    """
    scaled_data = scaler.transform(df_features[numerical_columns])
    # Step 1: Create a DataFrame from scaled numerical data with correct column names
    df_scaled = pd.DataFrame(scaled_data, columns=numerical_columns)
    
    # Step 2: Concatenate scaled numerical data with the original categorical columns
    df_scaled_final = pd.concat([df_scaled, df_features[categorical_columns].reset_index(drop=True)], axis=1)
    
    return df_scaled_final
```
